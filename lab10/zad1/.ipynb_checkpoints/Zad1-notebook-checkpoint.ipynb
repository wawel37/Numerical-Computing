```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageOps
```

```python
def showImage(image):
    plt.rcParams["figure.figsize"] = (16,9)
    plt.imshow(image, cmap='gray')
    plt.show()
```

```python
galia = 255 - np.asarray(Image.open("lab10_galia.png").convert("L"))
showImage(galia)
galiaFreq = np.fft.fft2(galia)
```

```python
fish = np.asarray(Image.open("lab10_school.jpg").convert("L"))
showImage(fish)
fishFreq = np.fft.fft2(fish)
```

```python
fishSearched = np.asarray(Image.open("lab10_fish1.png").convert("L"))
showImage(fishSearched)
```

```python
galiaSearchedFreq = np.fft.fft2(np.rot90(galiaSearched, 2), s=galia.shape)
fishSearchedFreq = np.fft.fft2(np.rot90(fishSearched, 2), s=fish.shape)
```

```python
def draw3DPlot(corelation):
    X = list(range(corelation.shape[1]))
    Y = list(range(corelation.shape[0]))
    X, Y = np.meshgrid(X, Y)
    
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    ax.plot_surface(X, Y, np.real(corelation), rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

```python
galiaCorelation = abs(np.fft.ifft2(np.multiply(galiaFreq, galiaSearchedFreq)))
fishCorelation = abs(np.fft.ifft2(np.multiply(fishFreq, fishSearchedFreq)))

draw3DPlot(galiaCorelation)
draw3DPlot(fishCorelation)

galiaMax = np.amax(galiaCorelation)
fishMax = np.amax(fishCorelation)
```

```python
showImage(fishCorelation)
showImage(galiaCorelation)
```

```python
fishOG = np.array(Image.open("lab10_school.jpg").convert("RGB"))
galiaOG = np.array(Image.open("lab10_galia.png").convert("RGB"))
```

```python
def getNearMaxArray(corelation, corelationMax, margin = 0.9):
    for i in range(corelation.shape[0]):
        for j in range(corelation.shape[1]):
            if corelation[i][j] < corelationMax * margin:
                corelation[i][j] = 0
    return corelation
```

```python
galiaCorelation = getNearMaxArray(galiaCorelation, galiaMax)
fishCorelation = getNearMaxArray(fishCorelation, fishMax, 0.7)
```

```python
for i in range(galiaCorelation.shape[0]):
    for j in range(galiaCorelation.shape[1]):
        if galiaCorelation[i][j] != 0:
            for width in range(galiaSearched.shape[0]):
                galiaOG[i - width, j] = (255,0,0)
                galiaOG[i - width, j - galiaSearched.shape[1]] = (255,0,0)
            for height in range(galiaSearched.shape[1]):
                galiaOG[i, j - height] = (255,0,0)
                galiaOG[i - galiaSearched.shape[0], j - height] = (255,0,0)

for i in range(fishCorelation.shape[0]):
    for j in range(fishCorelation.shape[1]):
        if fishCorelation[i][j] != 0:
            for width in range(fishSearched.shape[0]):
                fishOG[i - width, j] = (255,0,0)
                fishOG[i - width, j - fishSearched.shape[1]] = (255,0,0)
            for height in range(fishSearched.shape[1]):
                fishOG[i, j - height] = (255,0,0)
                fishOG[i - fishSearched.shape[0], j - height] = (255,0,0)    
    
```
